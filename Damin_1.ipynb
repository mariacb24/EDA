```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/My Drive/Data Mining/dataset.xls'

df = pd.read_excel(file_path)

print(df.head())
```

```python
mode_institusi = df['institusi'].mode()[0]
df['institusi'] = df['institusi'].fillna(mode_institusi)

df
```

```python
mode_area = df['area'].mode()[0]
df['area'] = df['area'].fillna(mode_area)

df
```

```python
mode_rumah = df['rumah'].mode()[0]
df['rumah'].fillna(mode_rumah, inplace=True)

df
```

```python
mode_tipe = df['tipe'].mode()[0]
df['tipe'].fillna(mode_tipe, inplace=True)

df
```

```python
def categorize_biaya(biaya):
    if biaya <= df.biaya_berobat.quantile(0.25):
        return '<100.000'
    elif biaya <= df.biaya_berobat.quantile(0.50):
        return '500.001 - 750.000'
    elif biaya <= df.biaya_berobat.quantile(0.75):
        return '750.001 - 1.000.000'
    elif biaya <= df.biaya_berobat.quantile(1):
        return '1.000.001 - 2.000.000'
    else:
        return '2.000.001 - 6.000.000'

# Apply the function to create a new column 'biaya_berobat_class'
df['biaya_berobat_class'] = df['biaya_berobat'].apply(categorize_biaya)

df
```

```python
df = df.drop('biaya_berobat', axis=1)

df
```

```python
import numpy as np

def find_outliers_iqr(data):
  Q1 = np.percentile(data, 25)
  Q3 = np.percentile(data, 75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = [x for x in data if x < lower_bound or x > upper_bound]
  return outliers

for column in dataolahencoding.columns:
  outliers = find_outliers_iqr(dataolahencoding[column])
  if outliers:
    print(f"Kolom '{column}' memiliki outlier:")
    print(outliers)
    print()
```

```python
import seaborn as sns

sns.pairplot(dataolahencoding)
plt.title('Pairplot of Features')
plt.show()
```

```python
x = dataolahencoding
pca = PCA(n_components=2)

principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])

print(principalDf)

plt.figure(figsize=(8, 6))
plt.scatter(principalDf['principal component 1'], principalDf['principal component 2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Hasil PCA')
plt.show()
```

```python
from sklearn.cluster import KMeans
n_clusters = 2
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(principalDf)
labels = kmeans.labels_
principalDf['cluster'] = labels

plt.figure(figsize=(8, 6))
for i in range(n_clusters):
  plt.scatter(principalDf[principalDf['cluster'] == i]['principal component 1'],
              principalDf[principalDf['cluster'] == i]['principal component 2'],
              label=f'Cluster {i}')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering Results')
plt.legend()
plt.show()
```

```python
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(principalDf)
labels = kmeans.labels_
principalDf['cluster'] = labels

plt.figure(figsize=(8, 6))
for i in range(n_clusters):
  plt.scatter(principalDf[principalDf['cluster'] == i]['principal component 1'],
              principalDf[principalDf['cluster'] == i]['principal component 2'],
              label=f'Cluster {i}')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering Results')
plt.legend()
plt.show()
```

```python
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(principalDf)
labels = kmeans.labels_
principalDf['cluster'] = labels

plt.figure(figsize=(8, 6))
for i in range(n_clusters):
  plt.scatter(principalDf[principalDf['cluster'] == i]['principal component 1'],
              principalDf[principalDf['cluster'] == i]['principal component 2'],
              label=f'Cluster {i}')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering Results')
plt.legend()
plt.show()
```

```python
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(principalDf[['principal component 1', 'principal component 2']])
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

# Based on the Elbow method, choose the optimal k value.
# Let's say the optimal k is 4 (you'll need to determine this from the plot).

optimal_k = 4  # Replace with the optimal k from the elbow method

kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(principalDf[['principal component 1', 'principal component 2']])
labels = kmeans.labels_
principalDf['cluster'] = labels

plt.figure(figsize=(8, 6))
for i in range(optimal_k):
    plt.scatter(principalDf[principalDf['cluster'] == i]['principal component 1'],
                principalDf[principalDf['cluster'] == i]['principal component 2'],
                label=f'Cluster {i}')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title(f'K-Means Clustering Results (Optimal k = {optimal_k})')
plt.legend()
plt.show()

```
