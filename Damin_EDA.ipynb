```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

# Replace 'path/to/your/dataset.xls' with the actual path to your Excel file in Google Drive
file_path = '/content/drive/My Drive/Data Mining/dataset.xls'

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(file_path)

# Now you can work with the data in the DataFrame 'df'
print(df.head())
```

```python
# Find the mode of the 'institusi' column
mode_institusi = df['institusi'].mode()[0]

# Fill NaN values in 'institusi' with the mode
df['institusi'] = df['institusi'].fillna(mode_institusi)

df
```

```python
# Find the mode of the 'area' column
mode_area = df['area'].mode()[0]

# Fill NaN values in 'area' with the mode
df['area'] = df['area'].fillna(mode_area)

df
```

```python
mode_rumah = df['rumah'].mode()[0]

# Fill NaN values in 'rumah' with the mode
df['rumah'].fillna(mode_rumah, inplace=True)

df
```

```python
mode_tipe = df['tipe'].mode()[0]

# Fill NaN values in 'tipe' with the mode
df['tipe'].fillna(mode_tipe, inplace=True)

df
```

```python
# Create a function to categorize the 'biaya_berobat' column into quantiles
def categorize_biaya(biaya):
    if biaya <= df.biaya_berobat.quantile(0.25):
        return '<100.000'
    elif biaya <= df.biaya_berobat.quantile(0.50):
        return '500.001 - 750.000'
    elif biaya <= df.biaya_berobat.quantile(0.75):
        return '750.001 - 1.000.000'
    elif biaya <= df.biaya_berobat.quantile(1):
        return '1.000.001 - 2.000.000'
    else:
        return '2.000.001 - 6.000.000'

# Apply the function to create a new column 'biaya_berobat_class'
df['biaya_berobat_class'] = df['biaya_berobat'].apply(categorize_biaya)

df
```

```python
# Drop the 'biaya_berobat' column
df = df.drop('biaya_berobat', axis=1)

df
```
